```python
from skimage import feature
from skimage.morphology import square
from skimage.filters import rank
def get_connectivity(a_slice):
    connectivity = rank.sum(a_slice.astype(bool).astype(int),square(3))
    return connectivity
    
```

```python
def iterate_forwards(image,number_of_iterations = 1):
    for i in range(number_of_iterations)
        connectivity = get_connectivity(image)
        probMatrix = 0.2*connectivity/9 + 0.8
        randMatrix = np.random.rand(probMatrix.shape[0],probMatrix.shape[1])
        sumMatrix = probMatrix - randMatrix
        image[np.where(sumMatrix<0)] = 0
    return image
    
show_slice(img[100:200,150:300,sl])
new_img = iterate(img[100:200,150:300,sl])
show_slice(new_img)

    
    
```
