```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import nibabel as nib
import numpy as np

all_scans_data = np.zeros((410,410,120,6))
tumour_file = "D:/fullscans/N01"
count = 0

for file in os.listdir("D:/fullscans/N01"):
    if file.rfind("a.nii") != -1:
        all_scans_data[:,:,:,count] = nib.load(os.path.join(tumour_file,file)).get_data()
        print(count)
        count += 1

 
```
