```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_table("D:/fractal-info/real_data_fractal_info.txt")
print(df.columns[86])
mode = df[df.columns[87]].mode()[0]
#df.where(df[df.columns[87]]>mode)[df.columns[5]].plot()
df[df.columns[87]].plot()
plt.show()

print(df.columns[0],df.columns[5],df.columns[86])
small_df = pd.DataFrame()
small_df["Image Name"] = df[df.columns[0]]
small_df["Fractal Dim"] = df[df.columns[5]]
small_df["Lacunarity"] = df[df.columns[86]]
small_df.head()
```

```python
def get_patient(x):
    x = x[0:x.find("_")]
    return x

def get_scan_no(x):
    x = x[x.find("_")+1:]
    x = x[:x.find("_")]
    return int(x)

def get_slice_no(x):
    x = x[x.find("_")+1:]
    x = x[x.find("_")+1:]
    x = x[:x.find("p")]
    return int(x)

small_df["Patient"] = small_df["Image Name"].apply(get_patient)

small_df["Scan No"] = small_df["Image Name"].apply(get_scan_no)

small_df["Slice No"] = small_df["Image Name"].apply(get_slice_no)

small_df.drop([small_df.columns[0]],axis = 1)
```

```python
file = open("D:/fractal-info/real_data_fractal_info_condensed.csv","w")
file.write("Patient, Scan No, Mean Fractal Dimension, Reduced Mean Fractal Dimension, Lacunarity, Reduced Lacunarity\n")
for patient in small_df["Patient"].unique():
    patient_df = small_df[small_df["Patient"] == patient]
    for scan in patient_df["Scan No"].unique():
        scan_df = patient_df[patient_df["Scan No"]==scan]
        fractal_dims = np.array(scan_df["Fractal Dim"])
        lacunarities = np.array(scan_df["Lacunarity"])
        reduced_fractal_dims = fractal_dims[fractal_dims != 1.8992]
        reduced_lacunarities = lacunarities[lacunarities != 0]
        file.write("%s,%s,%s,%s,%s,%s\n" % (patient,str(scan),str(fractal_dims.mean()),
                                            str(reduced_fractal_dims.mean()),str(lacunarities.mean()),
                                            str(reduced_lacunarities.mean())))
       
file.close()
```
